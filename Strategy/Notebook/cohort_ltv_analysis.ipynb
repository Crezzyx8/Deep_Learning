```python
#Synthetic Transactions 
import numpy as np
import pandas as pd

np.random.seed(42)
n = 3000

df = pd.DataFrame({
    "customer_id": np.random.randint(0, 800, n),
    "order_date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, n), unit="D"),
    "revenue": np.random.uniform(20, 200, n)
})

df["cohort"] = df.groupby("customer_id")["order_date"].transform("min").dt.to_period("M")
df["order_month"] = df["order_date"].dt.to_period("M")

cohort_data = df.groupby(["cohort", "order_month"])["revenue"].sum().reset_index()
cohort_data.head()
cohort_pivot = cohort_data.pivot(index="cohort", columns="order_month", values="revenue")
cohort_pivot.fillna(0, inplace=True)
cohort_pivot
```
