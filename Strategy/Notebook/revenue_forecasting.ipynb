# Revenue Forecasting

## Business Context
Accurate revenue forecasting supports budgeting and planning.

## Objective
Forecast future revenue using time series modeling.


```python
import pandas as pd
import numpy as np

dates = pd.date_range(start="2022-01-01", periods=800, freq="D")

revenue = (
    1000 +
    np.arange(800) * 2 +
    np.random.normal(0, 50, 800)
)

df = pd.DataFrame({"date": dates, "revenue": revenue})
df.head()
```

```python
df["ma_30"] = df["revenue"].rolling(30).mean()

import matplotlib.pyplot as plt

plt.plot(df["date"], df["revenue"], label="Actual")
plt.plot(df["date"], df["ma_30"], label="MA(30)")
plt.legend()
plt.title("Revenue Forecast Baseline")
plt.show()
```
