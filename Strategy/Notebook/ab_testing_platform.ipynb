```python
#Synthetic Experimental Data Generation and t-test
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

np.random.seed(42)

control = np.random.normal(100, 10, 1000)
treatment = np.random.normal(103, 10, 1000)

t_stat, p_val = ttest_ind(control, treatment)

print("Control Mean:", control.mean())
print("Treatment Mean:", treatment.mean())
print("p-value:", p_val)
```
